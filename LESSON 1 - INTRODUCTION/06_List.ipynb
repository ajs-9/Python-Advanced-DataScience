#### List

A list is an ordered, changeable collection of items, written with square brackets [].  
It can store multiple values of different data types.

```python
x = [16,27,43,"You"]
print(x)
```

```python
type(x)
```

```python
print(type(x))
```

```python
fruits = ["apple", "banana", "cherry"] 
```

Indexing - Start from 0 .., - 0,1,2,3,...

```python
print(fruits[0])   # apple (first item, index 0)
print(fruits[1])   # banana (second item, index 1)
print(fruits[-1])  # cherry (last item, using negative index)
print(fruits[-2])  #negavtive indexing
```

Slicing - Accessing elements in lists with the help of index numbers/ indicies, +ve = 0,1,2,3... or  -ve = -1,-2,-3....

```python
print(fruits[0:2])
print(fruits[:3])
```
