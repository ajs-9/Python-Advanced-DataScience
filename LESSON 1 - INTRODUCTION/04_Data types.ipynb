#### Data Types

Data type defines the kind of value a variable can hold.

#### Common Data Types:

- int → whole numbers
- float → decimal numbers
- str → text/string
- bool → True or False
- list → ordered, changeable collection
- tuple → ordered, unchangeable collection
- dict → key-value pairs
- set → unordered, unique values

#### Type Function

It is used to return the data type of an object

```python
a = 10
type(a)
```

```python
b = 10.3
type (b)
```

```python
c = "hello"
type(c)
```

```python
d = True
type(d)
```

```python
e = [1, 2, 3]
type(e)
```

```python
f = (1, 2, 3)
type(f)
```

```python
g = {"key": "value"}
type(g)
print(type(g))
```

```python
h = {1, 2, 3}
print(type(h))
```
