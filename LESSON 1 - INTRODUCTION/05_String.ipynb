#### String

String is a data type in python and a sequence of characters enclosed in quotes.

```python
a = 'harry' # Single quoted string
b = "harry" # Double quoted string
c = '''harry''' # Triple quoted string
```

Lenght of String

```python
len(a)
len(b)
len(c)
```

Indexing

```python
a = "Hello"
b = "World"
```

```python
print(a[0])
print(b[2])
```

Slicing

```python
print(a[1:4])
print(b[2:])    
```

Concatenation

```python
print(a + " " + b)
```
